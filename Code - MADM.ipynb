```python
import pandas as pd; import numpy as np
# Step: Drop columns
df_final = df3.drop(columns=['handwashing_facilities'])
df_final.replace(0, np.nan, inplace=True)

# Step: Replace missing values
df_final = df_final.fillna(df3.median())

df_final
```

# Alpha

```python
import pandas as pd; import numpy as np
# Step: Keep rows where variant is one of: Alpha
df2 = df_final.loc[df_final['variant'].isin(['Alpha'])]
pos_list = ['number_detections_variant', 'reproduction_rate', 'icu_patients', 'hosp_patients',
       'weekly_icu_admissions',
'weekly_hosp_admissions', 'positive_rate','excess_mortality', 'population',
'male_smokers',
       'female_smokers', 'diabetes_prevalence', 'cardiovasc_death_rate','extreme_poverty',
            'aged_65_older', 'aged_70_older',
       'median_age', 'population_density']
                       
neg_list=['new_tests', 'total_vaccinations','people_vaccinated', 'people_fully_vaccinated',
       'stringency_index','human_development_index', 'life_expectancy', 'gdp_per_capita']    
    
for col in pos_list:
    df2[col] = (df2[col] - df2[col].min()) / (df2[col].max() - df2[col].min())

for col in neg_list:
    df2[col] = (df2[col] - df2[col].max()) / (df2[col].min() - df2[col].max())    
    
corr_att = df2.corr()
std_att = df2.std()
C = std_att * ((1-corr_att).sum())
w = C/(C.sum())
S = (df2*w).sum(axis=1)
P = (df2**w).sum(axis=1)
ka = (P + S)/(P.sum() + S.sum())
kb = S/S.min() + P/P.min()
lamda = 0.5
kc = (lamda*S + (1-lamda)*P)/(lamda*S.max() + (1-lamda)*P.max())
final = (ka*kb*kc)**(1/3) + (1/3)*(ka+kb+kc)
final
Rank1 = pd.DataFrame({'Country': df2['country'],'Score': final })
Rank1 = Rank1.sort_values(by=['Country'], ascending=[True])

Rank1= Rank1.reset_index()  
Rank1 = Rank1.drop(columns=['index'])
Rank1 
    
```
